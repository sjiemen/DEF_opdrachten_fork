## *Opdracht 6*: Meten en fitten, Leerdoel

```python
# Zet hier jouw data in een voor jou handig formaat
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
Lux_phyphox = np.array([1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6,6.5,7,7.5,8,8.5,9,9.5,10])
Bits_arduino = np.array([59,65,72,74,79,83,87,94,103,106,108,112,115,126,130,136,137,141,149])

def function(x,a,b):
    return a*x +b
values, covariance = curve_fit(function,Bits_arduino,Lux_phyphox)
x_fit = np.linspace(55,150,1000)
y_fit = function(x_fit,*values)


err_bits = 1.5 # error van gemeten bits
err_lux = np.abs(function(Bits_arduino,*values) - Lux_phyphox)  # error van gemeten lux

plt.figure()
plt.plot(Bits_arduino,Lux_phyphox,'b.')
plt.errorbar(Bits_arduino,Lux_phyphox,err_lux,err_bits, fmt = 'none')
plt.plot(x_fit,y_fit,'r-')
plt.title('gemeten lux uitgezet tegenover gemeten bits')
plt.xlabel('lichtsterkte (bits)')
plt.ylabel('licht (lux)')
plt.show()

```

## *Opdracht 8*: Conclusie, leerdoel

Conclusie: 
Het ontwerp voldoet aan de opgestelde criteria. De errorbars van onze metingen zijn kleiner dan de stapgrootte, wat betekent dat we nauwkeurig genoeg de lichtintensiteit kunnen meten met een stapgrootte van 1 lux. Dit betekent dat het kan worden toegepast op een LED dat zal dimmen per afname van 1 lux, vanaf 10 lux tot 0 lux.
